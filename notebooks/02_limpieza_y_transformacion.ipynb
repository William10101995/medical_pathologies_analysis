## Importo librerias y cargo el dataset original

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/raw/diabetes.csv')
```

## Quito filas duplicadas

```python
df = df.drop_duplicates()
```

## Imputación de valores nulos

Teniendo en cuenta lo analizado en el archivo de exploración, rellenaremos los valores nulos con valores nulos de las columnas que correspondan. Esto lo haremos aplicando la mediana y agrupando de la manera mas homogenea posible segun la consideración del equipo (*gender y smoking_status*).

```python
cols_to_impute = ['age', 'bmi', 'blood_pressure', 'glucose_levels']

for col in cols_to_impute:
    df[col] = df.groupby(['gender', 'smoking_status'])[col].transform(
        lambda x: x.fillna(x.median())
    )
```

## Normalizamos columnas importantes

```python
df['gender'] = df['gender'].str.strip().str.lower()
df['smoking_status'] = df['smoking_status'].str.strip().str.lower()
```

## Corregimos tipos 

```python
df['age'] = df['age'].astype('int')
```

## Exporto el dataset limpio

```python
df.to_csv('../data/cleaned/diabetes_cleaned.csv', index=False)
```
